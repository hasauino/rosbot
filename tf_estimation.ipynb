```octave
sys =tf([1],[1 2 5]);
t = 0:0.001:2.999;

u1 = sin(2*pi*t*1);
u2 = sin(2*pi*t*2);
u3 = sin(2*pi*t*3);
u4 = sin(2*pi*t*4);
u5 = sin(2*pi*t*5);
u6 = sin(2*pi*t*6);
u7 = sin(2*pi*t*7);
u8 = sin(2*pi*t*8);

u = [u1 u2 u3 u4 u5 u6 u7 u8];
t = 0:0.001:23.999;
y = lsim(sys,u,t);
```

```octave
plot(t,u, t,y)
```

```octave
syse = tfestimate(u, y);
```
